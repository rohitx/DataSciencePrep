## Accessing Text from the Web and from Disk ##

We will begin by downloading a book *Crime and Punishment* over the internet. 

```python
import nltk, re, pprint
```

```python
# Let's get the book 
from urllib import urlopen
url = "http://www.gutenberg.org/files/2554/2554.txt"
raw = urlopen(url).read()
print "Data type :", type(raw)
print "Length of Data: ", len(raw)
print raw[:75]
```

To analyze the book we want to break up the string into words and punctuations. This process is called **tokenization**. The process produces a list with each element a string, something we have seen in previous two chapters. 

```python
tokens = nltk.word_tokenize(raw)
print len(tokens)
tokens[1:10]
```

In order to use all of the linguistic processing we learned in previous chapters we need to convert the text in NLTK text. 

```python
text = nltk.Text(tokens)
print type(text)
text[1020:1025]
```

```python
# We can also apply the collocation function 
text.collocations()
```

Notice that the collocation also has the words Gutenberg. Then appear because the text contains the names of people, companies, licensing companies that have worked on the text. It is hard to know what would be out in the text. We can however, remove them by finding strings. Of course, we need to have some knowledge of how the strings look like but once we know we can use the **find()** and **rfind()** functions to find such text

```python
raw.find("PART II")
```

```python
raw.rfind("End of Project Gutenberg's Crime")
```

So we have the indices of Part II and end of the book. We can now slice the book so that it begins with part II. 

```python
raw_book = raw[201099:1157746]
raw_book[:50]
```

So we begin with Part II, Chapter 1. The **rfind()** is the *reverse find* function that finds from the bottom up.

These steps are something we need to follow when working with text. 

## Dealing with HTML ##

Much of the text that exists on the internet is composed of HTML. Let's learn how to deal with HTML tags. 

```python
url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
html = urlopen(url).read()
html[:60]
```

Removing HTML tags out of webpages is so routine that that there exists a Python library called BeautifulSoup. To extract the html, use BeautifulSoup() function and to extract just the text use the methods .get_text(). Finally, tokenize the raw text and viola, we have our data.  

```python
from bs4 import BeautifulSoup
url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
html_doc = urlopen(url).read()
soup = BeautifulSoup(html_doc)
raw = soup.get_text()
tokens = nltk.word_tokenize(raw)
BBC_text = nltk.Text(tokens)
BBC_text[:10]
```

```python
# We can use concordance to find where a certain word has occured 
BBC_text.concordance('blonde')
```

## Regular Expressions for Detecting Word Patterns ##

We have seen previously in Chapter 1 that we can use Python in-built functions for Strings such as .endswith() or .startswith() that do pattern matching. Python has a powerful package called "regular expression" that give us powerful and flexible methods for describing the character patterns we are interested in. 

To use the regular expression package in Python we need to import it. 

```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

## Using Basic Metacharacters ##

Let's find words that end with "ed". The way to search this is by using a wildcard $. The $ sign at the end, like "ed$" signifies that we are looking for words ending with "ed". The search function is **re.search(a,b)** where a is the expression and b is the word. 

```python
ed_words = [w for w in wordlist if re.search('ed$', w)]
ed_words[:5]
```

Suppose we want to find a word that has a j in the third position and t in the sixth position and the word has 8 letters. We can find this with the wildcard **.**. The dot wild card for this search would look like this:

```python
jt_words = [w for w in wordlist if re.search('^..j..t..$', w)]
jt_words[:5]
```

The **caret symbol** matches the start of the string, just like $ matches the end. In the example, the caret symbol states that after the beginning of the word, the third position is a j and 6th position is a t. If we do not include the caret symbol, then we will find words that has j anywhere before t while t is exactly 2 positions before the end of the word. 

Finally, the **question symbol**, ?, specifies that the previous character is optional. This ^e-?mail$ will return both email and e-mail. 

## Ranges and Closures ##

We can query words like we would type them on a telephone pad. For example, the following query will result in:

```python
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
```

the query states that the word should start with a 'g', 'h', or 'i' and wnd with 'd', 'e', and 'f'. The basic regular expression metacharacters including wildcards, ranges and closures are given in the table below.
