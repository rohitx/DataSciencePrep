##Brown Corpus ##

The Brown Corpus was the first million-word electronic corpus of English, created in 1961 at Brown University. The corpus contains text from 500 sources, and have been categories by genre. The genre contains news, reviews, religion, fiction, mystery, government, etc...

We can access the corups as a list of words or a list of sentences (where each sentence is itself just a list of words). 

```python
from nltk.corpus import brown
brown.categories()
```

```python
# Let's look at works from the category "News"
brown.words(categories="news")
```

Let's compare genres in their usage of modal verbs, that is verbs pertaining to the mood. 

```python
news_text = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print m + ':', fdist[m]
```

**Exercise:** Choose a different section of the Brown Corpus, and adapt the preceding example to count a selection of wh words. 

```python
sci_fi = brown.words(categories="science_fiction")
fdist = nltk.FreqDist([w for w in sci_fi if w.lower().startswith('wh')])
this_modals = ['what', 'when', 'where', 'who', 'why']
for m in this_modals:
    print m + ":", fdist[m]
```

```python
# We can check this across all the genres
cfd = nltk.ConditionalFreqDist(
                    (genre, word)
                     for genre in brown.categories()
                     for word in brown.words(categories=genre))

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

## Conditional Frequency ##

As we saw earlier **FreqDist()** computes the number of occurances of each item in the list. A **conditional frequency distribution** is a collection of frequency distributions, each one for a different "condition". The condition will often filter by the category of the text. Conditional frequency allows us to maintain separate frequency distributions for each category. 

A frequency distribution counts observable events, such as the appearance of words in a text. A conditional frequency distribution needs to pair each event with a condition. So instead of processing a sequence of words, we have to process a sequence of pairs. 

### Counting Words by Genre ###

We used conditional frequency on the Brown Corpus to look for the frequency of certain words among all genre. Let's do the same but this time look for words among two genres. 

```python
genre_word = [(genre, word)
              for genre in ["news", "romance"]
              for word in brown.words(categories=genre)]
```

```python
len(genre_word)
```

```python
genre_word[:4]
```

```python
genre_word[-4:]
```

Think of it in this way. If you have positiv feedbacks and negative feedbacks, you can consider them as genre. You can then tag each word as a particular type. You can use this data to train a Naives Bayes classifier. 

We can now use this list of pairs to create a **ConditionalFreqDist** and save it in a variable cfd. As usual, we can type the variable to inspect it, and verify it has two conditions

```python
cfd = nltk.ConditionalFreqDist(genre_word)
cfd.conditions()
```

```python
# Let's access the two conditions and satify ourselves that each is just 
# a frequency distribution
cfd['news']
```

```python
cfd['romance']
```

```python
cfd['romance']['could']
```

```python
cfd['news']['could']
```

## Wordlist Corpora ##

NLTK includes some corpora that are nothing more than wordlists. The Words Corpus is a file used by some spellcheckers. We can use it to find unusual or misspelled words in a text corpus as follows

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusualWords= unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
print unusualWords[:4]
```

```python
(unusual_words(nltk.corpus.nps_chat.words()))[:5]
```

There is also a corpus of **stopwords**, that is, high-frequency words such as *the, to, also* that we sometimes want to filter out of a document before further processing. Stopwords have little lexical content, and their presence in a text fails to distinguish it from other texts. 

```python
from nltk.corpus import stopwords
len(stopwords.words('english'))
```

```python
(stopwords.words('english'))[:10]
```

```python
# This function computes the fraction of words in a text that are not in 
# the stopwords list:
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / float(len(text))
```

```python
content_fraction(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

## WordNet ##

Consider the following two sentences:

a. Benz is credited with the invention of the motorcar. 

b. Benz is credited with the invention of the automobile.

The two sentences are identical in the meaning and are different because we swapped motocar with automobile. We can explore synonyms with WordNet. 

```python
from nltk.corpus import wordnet as wn
wn.synsets('motorcar')
```

The result suggests that the word 'motorcar' has only one meaning. It is identical to the word car, which is a noun, and has one meaning. The entity **car.n.01** is called a synset or "synonymn set" a collection of synonymous words or **lemmas**. We can find the synonymns of words in the following way:

```python
wn.synset('car.n.01').lemma_names()
```

```python
wn.synset('car.n.01').definition()
```

```python
wn.synset('car.n.01').examples()
```

```python
# We can also look at the synonymns of the word car:
wn.synsets('car')
```

```python
# We see that there are 5 synonymns. We can print them as follows:
for sy in wn.synsets('car'):
    print sy.lemma_names()
```

```python
# Let's look at the word 'dish'
wn.synsets('dish')
```

```python
for sy in wn.synsets('dish'):
    print sy.lemma_names()
```

## The WordNet Hierarchy ##

WordNet makes it easy to navigate between concepts. For example, given a concept motorcar, we can look at the concepts that are more specific, the immediate ones. These are called **hyponyms**. 

```python
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
types_of_motorcar
```

## Lexical Relations ##

The components of the root word is called **meronyms**. For example, if the root word is 'tree', then the **part_meronyms()** will be trunk, crown and so on. The **substance_meronyms()** would be heartwood, sapwood. 

Things that are contained in is called **holonyms**. So the member holonym of a tree would be a forest. 

```python
wn.synset('tree.n.01').part_meronyms()
```

```python
wn.synset('tree.n.01').substance_meronyms()
```

```python
wn.synset('tree.n.01').member_holonyms()
```

root words that involve an action is found with the keyword **entailments()**. Finally, the lemmas that involve the opposite, the **antonyms()** as follows: 

```python
wn.synset('walk.v.01').entailments()
```

```python
wn.synset('eat.v.01').entailments()
```

```python
wn.lemma('rush.v.01.rush').antonyms()
```

```python
# Let's find the antonym of 'pride'
wn.synsets('pride')
```

```python
wn.lemma('pride.n.01.pride').antonyms()
```

Note that you need to use the root word twice in the argument of wn.lemma() before we can get the antonym of the word.
